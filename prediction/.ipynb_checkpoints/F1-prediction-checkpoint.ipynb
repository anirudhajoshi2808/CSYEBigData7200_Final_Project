```scala
import org.apache.spark.sql.SparkSession

val spark = SparkSession
  .builder()
  .appName("F1 Prediction")
  .getOrCreate()
```

```scala
val sc = spark.sparkContext
```

```scala
val drivers = spark.read.format("csv")
  .option("header", "true")
  .load("../data/drivers.csv")

val races = spark.read.format("csv")
  .option("header", "true")
  .load("../data/races.csv")

val quali = spark.read.format("csv")
  .option("header", "true")
  .load("../data/qualifying.csv")

val results = spark.read.format("csv")
  .option("header", "true")
  .load("../data/results.csv")

val constructors = spark.read.format("csv")
  .option("header", "true")
  .load("../data/constructors.csv")

// Smoothing constructor histories
val smoothedConstructors = constructors
  .withColumn("mergedId", when(col("constructorId") === 10, 211).otherwise(col("constructorId")))
  .withColumn("mergedId", when(col("constructorId") === 15, 51).otherwise(col("mergedId")))
```

```scala
val races2020 = races.filter($"year" === 2020).orderBy($"date")
val hungarianRace = races2020.filter($"name" === "Hungarian Grand Prix").select($"raceId")
val britishRace = races2020.filter($"name" === "British Grand Prix").select($"raceId")

val hungarian = results.filter($"raceId" === 1042)
  .withColumn("raceId", lit(1043))
  .withColumn("resultId", $"resultId" + 100000)
  .withColumn("grid", lit(0))
  .withColumn("position", lit(0))
  .withColumn("positionText", lit(""))
  .withColumn("positionOrder", lit(0))
  .select($"raceId", $"resultId", $"driverId", $"constructorId", $"number", $"grid", $"position", $"positionText", $"positionOrder")
  .withColumn("grid", when($"driverId" === 1, 1)
                        .when($"driverId" === 822, 2)
                        .when($"driverId" === 830, 7)
                        .when($"driverId" === 848, 13)
                        .when($"driverId" === 846, 8)
                        .when($"driverId" === 815, 4)
                        .when($"driverId" === 840, 3)
                        .when($"driverId" === 817, 11)
                        .when($"driverId" === 832, 9)
                        .when($"driverId" === 826, 17)
                        .when($"driverId" === 8, 20)
                        .when($"driverId" === 825, 16)
                        .when($"driverId" === 154, 18)
                        .when($"driverId" === 841, 19)
                        .when($"driverId" === 842, 10)
                        .when($"driverId" === 847, 12)
                        .when($"driverId" === 849, 15)
                        .when($"driverId" === 839, 14)
                        .when($"driverId" === 844, 6)
                        .when($"driverId" === 20, 5)
                        .otherwise($"grid"))
  .withColumn("position", when($"driverId" === 1, 1)
                            .when($"driverId" === 822, 3)
                            .when($"driverId" === 830, 2)
                            .when($"driverId" === 848, 5)
                            .when($"driverId" === 846, 13)
                            .when($"driverId" === 815, 7)
                            .when($"driverId" === 840, 4)
                            .when($"driverId" === 817, 8)
                            .when($"driverId" === 832, 9)
                            .when($"driverId" === 826, 12)
                            .when($"driverId" === 8, 15)
                            .when($"driverId" === 825, 10)
                            .when($"driverId" === 154, 16)
                            .when($"driverId" === 841, 17)
                            .when($"driverId" === 842, 20)
                            .when($"driverId" === 847, 18)
                            .when($"driverId" === 849, 19)
                            .when($"driverId" === 839, 14)
                            .when($"driverId" === 844, 6)
                            .when($"driverId" === 20, 5)
                            .otherwise($"position"))
```

```scala
val quali_hungary = quali.filter($"raceId" === 1042)
                        .selectExpr("qualifyId + 100000 as qualifyId", "1043 as raceId", "driverId", "constructorId", "number", "'' as q1", "'' as q2", "'' as q3")
val merged_quali_hungary = quali_hungary.join(hungarian.select("driverId", "grid"), Seq("driverId"), "left")
                                          .withColumnRenamed("grid", "position")
```

```scala
val british = results.filter($"raceId" === 1042)
                     .select($"raceId".alias("raceId"), 
                             ($"resultId" + 100050).alias("resultId"),
                             $"driverId", $"constructorId",
                             $"number", lit(0).alias("grid"),
                             lit(null).alias("position"),
                             lit("").alias("positionText"),
                             lit(null).alias("positionOrder"))

british.filter($"driverId" === 1).withColumn("grid", lit(1))
        .filter($"driverId" === 822).withColumn("grid", lit(2))
        .filter($"driverId" === 830).withColumn("grid", lit(3))
        .filter($"driverId" === 848).withColumn("grid", lit(12))
        .filter($"driverId" === 846).withColumn("grid", lit(5))
        .filter($"driverId" === 815).withColumn("driverId", lit(807))
                                   .withColumn("number", lit(27))
                                   .withColumn("grid", lit(13))
        .filter($"driverId" === 840).withColumn("grid", lit(6))
        .filter($"driverId" === 817).withColumn("grid", lit(8))
        .filter($"driverId" === 832).withColumn("grid", lit(7))
        .filter($"driverId" === 826).withColumn("grid", lit(19))
        .filter($"driverId" === 8).withColumn("grid", lit(16))
        .filter($"driverId" === 825).withColumn("grid", lit(14))
        .filter($"driverId" === 154).withColumn("grid", lit(17))
        .filter($"driverId" === 841).withColumn("grid", lit(15))
        .filter($"driverId" === 842).withColumn("grid", lit(11))
        .filter($"driverId" === 847).withColumn("grid", lit(20))
        .filter($"driverId" === 849).withColumn("grid", lit(18))
        .filter($"driverId" === 839).withColumn("grid", lit(9))
        .filter($"driverId" === 844).withColumn("grid", lit(4))
        .filter($"driverId" === 20).withColumn("grid", lit(10))
```

```scala
british.show(5)
```

```scala
val quali_british = quali.filter($"raceId" === 1042)
  .select(col("qualifyId") + 100050 as "qualifyId", lit(1044) as "raceId", $"driverId", $"constructorId", $"number", lit("") as "q1", lit("") as "q2", lit("") as "q3")
  .withColumn("driverId", when($"driverId" === 815, 807).otherwise($"driverId"))
  .join(british.select($"driverId", $"grid" as "position"), Seq("driverId"))
  .withColumn("position", when($"driverId".isin(847, 826), $"position" - 5).otherwise($"position"))
  .withColumn("position", when($"driverId".isin(8, 825, 154, 841, 849), $"position" + 2).otherwise($"position"))
```

```scala
val quali_british = quali.filter($"raceId" === 1042)
  .selectExpr(
    "qualifyId + 100050 as qualifyId",
    "1044 as raceId",
    "driverId",
    "constructorId",
    "number",
    "'' as q1",
    "'' as q2",
    "'' as q3"
  )

quali_british.filter($"driverId" === 815)
  .withColumn("driverId", lit(807))
  .withColumn("number", lit(27))

val british_df = british.select($"driverId", $"grid".as("position"))

quali_british.join(british_df, Seq("driverId"), "inner")
  .withColumn("position", when($"driverId".isin(847, 826), $"position" - 5).otherwise($"position"))
  .withColumn("position", when($"driverId".isin(8, 825, 154, 841, 849), $"position" + 2).otherwise($"position"))
```
